```python
import pandas as pd
from numpy import genfromtxt
import scipy.misc
from IPython.display import Image
```

```python
def view_image(csv_path):
    array_image = genfromtxt(csv_path, delimiter=';')
    jpg_path = csv_path + '.jpg'
    scipy.misc.imsave(jpg_path, array_image)
    return Image(jpg_path)
```
